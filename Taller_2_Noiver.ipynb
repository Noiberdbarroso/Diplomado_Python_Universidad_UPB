```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mp1

## cree una clase para modelar nuestros objeto 
class modelo_erizo():
    
    ## cree una lista tal cual como lo indica el taller 
    
    datospoints=[(1,1),(3,3),(5,5),(-1,-1)]
    possible_ms = []
    possible_bs = []
    
    # se realizo un metodo get_y() para calcular la pendiente 
    def get_y(self,valorm,valorb,valorx):
        valory=0
        valory=valorm*valorx+valorb
        
        return valory
    
    #relizamos una metodo que ayude calcular el error con los datos anteriosres 
    def calcular_error(self,valorm,valorb,point):
        x_point=point[0];
        y_point=point[1];
        
        y= self.get_y(valorm,valorb,x_point)
    
        diferencia= y-x_point        
        
        return abs(diferencia)
    ## para calcular el error con mas datos (datospoint) se realiza el metodo siguente para obtener un error mas pequeño  
    
    def calculate_all_error(self, valorm,valorb):
        erro_total=0
        
        for i in self.datospoints:
            
            erro_total = erro_total + self.calcular_error(valorm,valorb,i)
            
        return erro_total
    
    ## se creo una metodo que contiene 2 lista de posibles coon datos b, m desde -20 a 20 aumentandolo desde 0.1
    def listas_aproximada(self):
        
        self.possible_bs=[ inicio_c for inicio_c in np.arange(-20, 20.01, 0.1)]
        self.possible_ms=[ inicio_c for inicio_c in np.arange(-20, 20.01, 0.1)]
  
        
        
        ##  con este metodo se quiere buscar los datos que arrogen el error mas pequeño 
        
    def aproximacion_error(self):
        smallest_error = 100.0
        
        best_m=0.0
        best_b=0.0
        
        for punto_m in self.possible_ms:
            for punto_b in self.possible_bs:
                error_actual= self.calculate_all_error(punto_m,punto_b)
                
                if smallest_error > error_actual:
                    
                    smallest_error=error_actual
                    
                    best_m=punto_m
                    
                    best_b=punto_b
                    
                    
        
        return smallest_error,best_m,best_b
   
  
        
       
    
    ## creamos un objeto de la clase modelo_erizo 
erizo = modelo_erizo()
```

```python
## probamos los metodos ya antes creados 
print("datos de prueba (1,0(3,3))")
print (erizo.calcular_error(1,0,(3,3)))
print("datos de prueba (1,0(3,4))")
print (erizo.calcular_error(1,0,(3,4)))

print("datos de prueba (1,-1(3,3))")       
print (erizo.calcular_error(1,-1,(3,3)))
print("datos de prueba (-1,1(3,3))")
print (erizo.calcular_error(-1,1,(3,3)))

print("datos de prueba (1,0")
print (erizo.calculate_all_error(1,0))
print("datos de prueba (1,1")
print (erizo.calculate_all_error(1,1))
print("datos de prueba (1,-1")
print (erizo.calculate_all_error(1,-1))
print("datos de prueba (-1,1")
print (erizo.calculate_all_error(-1,1))

erizo.listas_aproximada()

```
